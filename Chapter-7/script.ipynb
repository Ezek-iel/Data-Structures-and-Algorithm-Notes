# Dijkstra's Algorithm

__Breadth first search__ is used to find a path with the fewest segments  
__Dijkstra's algorithm__ is used to look for the fastest path instead



## Steps to Dijkstra's Algorithm
1. Find the cheapest node (Neighbours with the lowest values)
2. Update the costs of the neighbours of the node
3. Repeat until you have done this for every node in the graph
4. Calculate the final path

> Remember to forget the former node

Also Graphs can also have cycles  
  


Also remember that an undirected graph contains cycles by default
  

  


> Note that Dijkstra's algorithm only works on directed acylic graphs or (DAG's  for short)
